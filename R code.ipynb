```r
install.packages('kohonen')
library(kohonen)
```

```r
mean_indexes <- read.csv(
    '/content/mean_indexes.csv', row.names='B_COUNTRY_ALPHA')
mean_indexes
```

```r
grid <- somgrid(xdim=6, ydim=6, topo='hexagonal', neighbourhood.fct='gaussian')
som <- som(scale(mean_indexes[, 1:8]), grid=grid, rlen=500)
```

```r
plot(som, type='changes')
```

```r
plot(som, type='counts', shape='straight')
```

```r
plot(som, type='mapping', labels=rownames(mean_indexes), shape='straight')
```

```r
plot(som, type='mapping', labels=mean_indexes$cluster, shape='straight')
```

```r
som.hc <- cutree(hclust(object.distances(som, type="codes")), k=9)
```

```r
plot(
  som, type='property', palette.name=rainbow, property=som.hc, shape='straight')
add.cluster.boundaries(som, som.hc)
```
